In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

Python has the following data types built-in by default, in these categories:

Text Type:	str
Numeric Types:	int, float, complex
Sequence Types:	list, tuple, range
Mapping Type:	dict
Set Types:	set, frozenset
Boolean Type:	bool
Binary Types:	bytes, bytearray, memoryview
None Type:	NoneType

```python
# Getting the Datatype
# You can get the datatype of any object using the type() function
x=5
print(type(x))
```

```python
# Setting the Datatype
# In Python, the datatype is set when you assign a value to a variable

x='Hello World'
y=20
z=20.5
a=["apple","banana","cherry"]
b=("apple","banana","cherry") 
c=range(6)
d={"name":"Rohit", "age":"27"}
e=frozenset({"apple","banana","cherry"})
f=True
g=b"Hello"
h=bytearray(5)
i=memoryview(bytes(5))
j=None 

#display x
print(x)
print(y)
print(z)
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h) 
print(i)
print(j) 

#display the datatype of x
print(type(x))
print(type(y))
print(type(z))
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
print(type(i))
print(type(j)) 
```

```python
# Setting the Specific Datatype
# If you want to specify the data type, you can use the following constructor functions:

x=str('Hello World!')
y=int(20)
z=float(20.5)
a=complex(1j)
b=list(("apple","cherry","banana"))
c=tuple(("apple","cherry","banana"))
d=range(6)
e=dict(name="John",age=36)
f=set(("apple","banana","cherry"))
g=frozenset(("apple","banana","cherry"))
h=bool(5)
i=bytes(5)
j=bytearray(5)
k=memoryview(bytes(5))



print(x)
print(y)
print(z)
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
print(i)
print(j)
print(k)

print(type(x))
print(type(y))
print(type(z))
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
print(type(i))
print(type(j))
print(type(k))
```
